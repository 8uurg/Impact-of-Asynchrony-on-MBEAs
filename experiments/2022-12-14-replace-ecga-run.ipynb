Due to a bug, a particular configuration failed to run successfully, resulting in an all-runs-failed assessment in the original run (`2022-12-13-scalability-ecga-trap-re-149d6ac-archive.csv.gz`).
We reran, this configuration, resulting in `2022-12-14-scalability-ecga-trap-re-2aa20e4-archive.csv.gz`. This notebook merges these two files, removing the original (bugged) sampled from the original run, replacing it with new samples from the rerun.

```python
import pandas as pd
from IPython.display import display
```

```python
# Load data
orig_runs = pd.read_csv("./results/2022-12-13-scalability-ecga-trap-re-149d6ac-archive.csv.gz")
reruns = pd.read_csv("./results/2022-12-14-scalability-ecga-trap-re-2aa20e4-archive.csv.gz")
out_path = "./results/2022-12-13-scalability-ecga-trap-re-149d6ac-2aa20e4-archive.csv.gz"
display(orig_runs)
display(reruns)
```

```python
# filter out the synchronous generational runs
orig_runs = orig_runs[(orig_runs["replacement_strategy"] != 6) | (orig_runs["algorithm_type"] != "sync")]
```

```python
# combine
all_runs = pd.concat([orig_runs, reruns])
```

```python
# save
all_runs.to_csv(out_path)
```
